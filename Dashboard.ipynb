```python
import plotly.express as px
import dash
from dash import dcc
from dash import html
import plotly.graph_objects as go
from dash.dependencies import Input, Output
```

```python
univer=['Massachusetts Institute of Technology (MIT) ','University of Cambridge','Stanford University','University of Oxford','Harvard University','California Institute of Technology (Caltech)','Imperial College London','UCL','ETH Zurich - Swiss Federal Institute of Technology','University of Chicago','National University of Singapore (NUS)','Peking University','University of Pennsylvania','Tsinghua University','The University of Edinburgh']
```

```python
# Define your app
app = dash.Dash(__name__)

# Define layout
app.layout = html.Div(children=[ 
    html.H1("Dashboard of universities", style={'textAlign': 'center', 'color': '#f5f6fa', 'font-size': 40}), 
    html.Div(children=[
        dcc.Graph(figure=px.bar(count_univer, x="location", y='number_of_institutions', title='Number of institutions from countries QS 2023').update_traces(marker_color=["#4CBB17","#4169E1", '#4169E1', '#4169E1', '#4169E1', '#4169E1', '#4169E1', '#4169E1', '#4169E1', '#7F00FF'])),
        dcc.Graph(figure=px.pie(top_6, values='sum', names=top_6.index, title="Nombre d'universités par continent"))
        ],
    style={'display': 'inline-flex', 'justify-content': 'space-around'}),
    dcc.Graph(figure=px.bar(mean_scored, x="location", y='mean_score_scaled', title='mean_score_scaled from countries QS 2023').update_traces(marker_color=["#4CBB17","#4169E1", '#4169E1', '#4169E1', '#4169E1', '#4169E1', '#4169E1', '#4169E1', '#4169E1', '#7F00FF'])),
    dcc.Dropdown(
        id='site-dropdown',  
        options=country_names,  
        value='Australia',  
        placeholder="Select a country here",  
        searchable=True,
        style={'width': '50%', 'margin': 'auto'}
    ),
    dcc.Graph(id='success-buble-chart'),
    
    dcc.RangeSlider(
        id='map-slider',
        min=0,
        max=15,
        step=1,
        value=[5, 15],
        marks={i: str(i) for i in range(0, 16)}
    ),
    dcc.Graph(id='map_plot'),
    
    dcc.Dropdown(
        id='univ-dropdown',  
        options=univer,  
        value='University of Cambridge',  
        placeholder="Select university here",  
        searchable=True,
        style={'width': '50%', 'margin': 'auto'}
    ),
    dcc.Graph(id='radar-chart'),
], style={'background-color': '#4169E1'})

# Callbacks
@app.callback(
    Output(component_id='success-buble-chart', component_property='figure'),
    Input(component_id='site-dropdown', component_property='value')
)
def update_scatter_chart(selected_country):
    selected_country_data = df1[df1["location"] == selected_country].sort_values(by='score scaled', ascending=False).head(20)
    fig = px.scatter(selected_country_data, x='institution', y='score scaled', 
                     title=f'Scatter Plot for best 20 universities in {selected_country}')
    return fig

@app.callback(
    Output(component_id='map_plot', component_property='figure'),
    Input(component_id='map-slider', component_property='value')
)
def update_map_plot(selected_top_range):
    lower_bound = selected_top_range[0]
    upper_bound = selected_top_range[1]
    
    traces=[]
    for index, row in df_for_map.iloc[lower_bound:upper_bound].iterrows():
        
        trace = go.Scattergeo(
            lat=[row['Latitude']],  # Latitude of the institution
            lon=[row['Longitude']],  # Longitude of the institution
            mode='markers',  # Set mode to 'markers'
            marker=dict(
                size=10,
                color='#4169E1',   #bleu ciel
                opacity=0.7,
                symbol='circle'),
            text=row['institution'],  # Institution name as text
            showlegend=False  # Hide the legend
        )
        traces.append(trace)
    
    layout = go.Layout(
        title=f'Map Visualization for {lower_bound,upper_bound}',
        geo=dict(
            projection_type="natural earth"#"mt flat polar parabolic"
        )
    )
    
    fig = go.Figure(data=traces, layout=layout)
    
    return fig

@app.callback(
    Output(component_id='radar-chart', component_property='figure'),
    Input(component_id='univ-dropdown', component_property='value')
)
def update_radar_chart(univ):
    sp_filtered = sp[sp["institution"] == univ]
            
    institution = sp_filtered['institution'].values[0]
    scores = sp_filtered.iloc[:, 1:].values.flatten().tolist()
    categories = sp_filtered.columns[1:].tolist()
    
    fig = go.Figure(data=go.Scatterpolar(
        r=scores,
        theta=categories,
        fill='toself',
        name=institution
    ))
    fig.update_layout(
        polar=dict(
            radialaxis=dict(visible=True),),
        showlegend=True
    )
    return fig  
    
if __name__ == '__main__':
    app.run_server()
```
